```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
prices_128 = []
prices_256 = []
with open("phones.csv", newline='') as file:
    csv_file = csv.reader(file, delimiter=";", quotechar='"')
    for line in csv_file:
        if '128' in line[2]:
            prices_128.append(int(line[0]))
        elif '256' in line[2]:
            prices_256.append(int(line[0]))
```

```python
print(f'Min: {np.array(prices_128).min()}')
print(f'Max: {np.array(prices_128).max()}')
print(f'Mean: {np.array(prices_128).mean()}')
```

```python
plt.plot(np.arange(1, 129),np.array(prices_128), color="#f27126", label="Prices 128 GB")
plt.plot(np.arange(1, 4),np.array(prices_256), label="Prices 256 GB")
plt.legend()
plt.show()
```

```python
plt.bar([1], [len(prices_128)], label="Quantity - 128 GB")
plt.bar([2], [len(prices_256)], label ="Quantity - 256 GB")
plt.legend()
plt.show()
```
